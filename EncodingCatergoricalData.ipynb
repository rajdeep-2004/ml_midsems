#Ordinal Encoding

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('customer.csv')
```

```python
df.sample(5)
```

```python
df.head()
```

```python
df = df.iloc[:, [2, 3, 4]]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop('purchased', axis=1),
                                                    df['purchased'],
                                                    test_size = 0.3,
                                                    random_state = 0)
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import OrdinalEncoder

o_encoder = OrdinalEncoder(categories=[['Poor', 'Average', 'Good'], ['School', 'UG', 'PG']])
# the list is in increasing order of importance , Poor = 0, Average = 1 and so on

o_encoder.fit(X_train)

X_train_o_encoded = o_encoder.transform(X_train)
X_test_o_encoded = o_encoder.transform(X_test)
```

```python
from sklearn.preprocessing import LabelEncoder

l_encoder = LabelEncoder()

l_encoder.fit(y_train)

y_train_l_encoded = l_encoder.transform(y_train)
y_test_l_encoded = l_encoder.transform(y_test)
```
